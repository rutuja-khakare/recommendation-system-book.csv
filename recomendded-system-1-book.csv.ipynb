```python
#import dataset
df=pd.read_csv("book.csv")
df
```

```python
#change data column name i.e replace '.'with '_'
df.columns = df.columns.str.replace('.', '_')
df
```

```python
# converting long data into wide data using pivot table
df1=df.pivot_table(index='User_ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
df1
```

```python
# Replacing the index values by unique user Ids
df1.index=df['User_ID'].unique()
df1
```

```python
#Impute those NaNs with 0 values
df1.fillna(0, inplace=True)
df1
```

# cosine similarity

```python
df2 = 1 - pairwise_distances( df1.values,metric='cosine') #pairwise dsitance for cosine
#1 in output represents similarity
df2
```

```python
#Store the results in a dataframe
df3 = pd.DataFrame(df2)
df3
```

```python
# Set the index and column names to user ids 
df3.index=df['User_ID'].unique()
df3.columns=df['User_ID'].unique()
df3
```
